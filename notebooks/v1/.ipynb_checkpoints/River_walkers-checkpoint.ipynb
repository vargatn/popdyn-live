```python
class WaterWalker(object):

    def __init__(self, hmap, hmap2 = None, startpos=None, wdraw=None, carry_fraction=0.1, maxes=(2400, 3200)):
        """water erosion walker"""

        self.path = []
        self.path_heights = []
        self.maxes = maxes
        
        self.wdraw = wdraw
        self.pos = startpos
        if self.pos is None:
            self.pos = self.get_startpos(self.wdraw)
        
        
        self.path.append(self.pos)
        self.hmap = hmap
        self.hmap2 = hmap2
        if self.hmap2 is None:
            self.hmap2 = np.zeros(hmap.shape)
        
        
        
        
        self.height = hmap[self.pos[0], self.pos[1]]
        self.path_heights.append(self.height)
        self.carry_height = 0
        
        self.carry_fraction = carry_fraction
        
    def step(self):
        _heights = []
        _poslist = []
        #find local direction
        for a in (-1, 0, 1):
            for b in (-1, 0, 1):
                i = self.pos[0] + a
                j = self.pos[1] + b
                if i > 0 and j > 0 and i < self.maxes[0] and j < self.maxes[1]:
                    _heights.append(self.hmap[i, j])
                    _poslist.append((i, j))
        inewh = np.argmin(_heights)

        self.newpos = _poslist[inewh]
        self.height = self.hmap[self.pos[0], self.pos[1]]
        self.new_height = self.hmap[self.newpos[0], self.newpos[1]]

        if self.new_height <= self.height:
            self.newpos = _poslist[inewh]
        else:
            return None
        
        
        self.carry_height = (self.height - self.new_height) *  self.carry_fraction        
        self.hmap2[self.pos[0], self.pos[1]] = self.carry_height
        #self.hmap2[self.pos[0], self.pos[1]] -= self.carry_height
        #self.hmap2[self.newpos[0], self.newpos[1]] += self.carry_height
        
        self.path.append(self.newpos)
        self.pos = self.newpos
        self.path_heights.append(self.height.copy())
        
    def walk(self, nstep=100):
        for i in np.arange(nstep):
            self.step()
                
        self.path = np.array(self.path)
        self.path_heights = np.array(self.path_heights)
        

    def _get_rand_pos(self):
        pos = [
            np.random.randint(low=0, high=self.maxes[0]),
            np.random.randint(low=0, high=self.maxes[1]),
        ]
        return pos
        
    def get_startpos(self, wland):
        pos = self._get_rand_pos()
        vol = wmap_land[startpos[0], startpos[1]]
        
        while vol < 1:
            pos = self._get_rand_pos()
            vol = wmap_land[startpos[0], startpos[1]]
            
        return pos
```

# a slew of walkers in series

```python
hmap_ep = hmap
for ep in np.arange(1):
    hmap2 = np.zeros(hmap.shape)
    for i in np.arange(10000):
        if i % 100 == 0:
            print(ep, i)
        WW = WaterWalker(hmap_ep, hmap2, carry_fraction=1)
        WW.walk(200)
        hmap2 = WW.hmap2
        
    tmp = cv.convolve_map(hmap2, 5)
        
    hmap_ep = hmap_ep - tmp
```
